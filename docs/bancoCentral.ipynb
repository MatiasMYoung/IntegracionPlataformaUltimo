```python
#SI SE EJECUTA DESPUES DE LAS 17:30, MUESTRA EL DÓLAR DEL SIGUIENTE DÍA HÁBIL

valor_dolar = siete.cuadro(
    series=["F073.TCO.PRE.Z.D"],
    nombres=["dolar"],
    desde=fecha_actual
)

# Obtener el último valor del dólar
valor_dolar = valor_dolar['dolar'].iloc[-1]

# Condicionar para que solo pueda ingresar números válidos
while True:
    try:
        numero = float(input("Precio en dólares: "))
        break
    except ValueError:
        print("Por favor, ingresa un número válido.")

# Multiplicar y luego truncar
transformacion = int(numero * valor_dolar)

# Separador de dígitos
transformacion_separado = f"{transformacion:,.0f}".replace(",", ".")

print(f"{numero} dólares ({valor_dolar}) = {transformacion_separado} CLP")
```
